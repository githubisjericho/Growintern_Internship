```python
# Handling Imbalanced Dataset
# Undersampling
normal = data[data['Class']==0]
fraud = data[data['Class']==1]
```

```python
normal.shape,fraud.shape
```

```python
normal_sample=normal.sample(n=473)
normal_sample.shape
```

```python
new_data = pd.concat([normal_sample,fraud],ignore_index=True)
new_data.head()
```

```python
X = new_data.drop('Class',axis=1)
y = new_data['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,
                                                 random_state=42)

# Logistic Regression
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred = log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score
```

```python
accuracy_score = accuracy_score(y_test,y_pred)
f1_score = f1_score(y_test,y_pred)

recall_score = recall_score(y_test,y_pred)

precision_score = precision_score(y_test,y_pred)


print(f"Accuracy Score of LR model is {accuracy_score}")

print(f"F1 Score of LR model is {f1_score}")

print(f"Recall Score of LR model is {recall_score}")

print(f"Precision Score of LR model is {precision_score}")
```
